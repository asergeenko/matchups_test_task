```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
outcome = pd.read_csv('outcome_interview.csv')
```

```python
outcome.head()
```

```python
event = pd.read_csv('event_interview.csv')
event.head()
```

```python
fighter = pd.read_csv('fighter_interview.csv')
fighter.head()
```

```python
outcome_copy = outcome.copy().drop(['id','winner_id'],axis=1)
new_matchups = outcome_copy.merge(fighter,left_on='fighter1_id',right_on='id',how='left').drop(['id','fighter1_id'], axis=1).merge(
    fighter,left_on='fighter2_id',right_on='id',how='left').drop(['id','fighter2_id'], axis=1).merge(
    event[['id','name']],left_on='event_id',right_on='id',how='left').drop(['id','event_id'],axis=1)

new_matchups.columns = ['fighter1_name','fighter1_nationality','fighter2_name','fighter2_nationality','event_name']
new_matchups = new_matchups[['fighter1_name','fighter2_name','fighter1_nationality','fighter2_nationality','event_name']]
new_matchups.head()
new_matchups.to_csv('new_matchups.csv',index=False)
```

```python
nationalities = new_matchups['fighter1_nationality'].append(new_matchups['fighter2_nationality']).value_counts(normalize=True)
row_cnt = len(nationalities.index)
threshold = 0.003
mask = nationalities > threshold
tail_prob = nationalities.loc[~mask].sum()
nationalities = nationalities.loc[mask]
nationalities['Other'] = tail_prob
explode = [0.01]*nationalities.size

def my_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 2 else ''

with plt.style.context({"axes.prop_cycle" : plt.cycler("color", plt.cm.tab20.colors)}):
    ax = nationalities.plot.pie(figsize=(10, 10),explode=explode,legend=True,fontsize=8,pctdistance=0.9,labeldistance=1.2,shadow=True,labels=None,autopct=my_autopct,
    counterclock=False).legend(nationalities.keys(),bbox_to_anchor=(1,1))
plt.title('Fighter Nationality Density Index')
plt.ylabel('')
plt.show()
```
